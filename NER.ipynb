```python
from mediawiki import MediaWiki
wikipedia = MediaWiki()

p = wikipedia.page('GT Nexus')
content= p.content
#print(content) #this is harvesting wikipedia data

import nltk
from nltk.tokenize import word_tokenize
tokenized_doc  = word_tokenize(content)
tagged_sentences = nltk.pos_tag(tokenized_doc)
NER= nltk.ne_chunk(tagged_sentences ) #here we have tokenized and chunked the sentences with NLTLK NER Tool


n_entity = []
for tagged_tree in NER:
    print(tagged_tree)
    if hasattr(tagged_tree, 'label'):
        entity_name = ' '.join(c[0] for c in tagged_tree.leaves()) #
        entity_type = tagged_tree.label() # get NE category
        n_entity.append((entity_name, entity_type))

    #print(n_entity)
    
for tag in n_entity:
    #print(tag[1])
    if tag[1]=='GPE':   #Specify any tag which is required
        print(tag) 
        #Location: America, Europe, Asia, Alameda County HQ: Oakland CA"
        #Names: 	Aaron Sasson, John Urban, Greg Johnsen, Anil Nair
        
        

        
```
